Гарантируется, что из знаков препинания в строке input_string могут содержаться только точки ('.'), запятые (','), восклицательные ('!') и вопросительные ('?') знаки.

```

## count_words = 13

input_string = 'There are many great articles about Artificial Intelligence and its benefits for business and society. However, many of these articles are too technical for the average reader.'

```

````python
input_string = 'Hello! My name is Python. I will help you to analyze some data.'

def countings(input_string):
   resultt = input_string.lower()
   resultt = resultt.replace('.', '')
   resultt = resultt.replace(',', '')
   resultt = resultt.replace('?', '')
   resultt = resultt.replace('!', '')
   result = len(resultt.split())
   return result

count_words = countings(input_string)
print(count_words)
````

В нашем распоряжении есть список строк разной длины.

Напишите функцию equalize_lengths(), которая принимает на вход список и возвращает новый список из строк одинаковой длины.

Длина итоговой строки должна быть равна длине самой длинной из строк. Если конкретная строка короче самой длинной, её необходимо дополнить до требуемого количества символов нижними подчёркиваниями с правого края.

Важно! Результирующий список должен быть отсортирован от меньшего к большему по количеству добавленных символов нижнего подчёркивания. То есть первым должен идти элемент с максимальной длиной строки.

Примеры работы программы:

```
print(equalize_lengths(['крот', 'белка', 'выхухоль']))
# ['выхухоль', 'белка___', 'крот____']

print(equalize_lengths(['a', 'aa', 'aaa', 'aaaa', 'aaaaa']))
# ['aaaaa', 'aaaa_', 'aaa__', 'aa___', 'a____']

print(equalize_lengths(['qweasdqweas', 'q', 'rteww', 'ewqqqqq']))

# ['qweasdqweas', 'ewqqqqq____', 'rteww______', 'q__________']
```

````python
def equalize_lengths(word_list):
   max_length = len(max(word_list, key=len))
   word_list = sorted(word_list, key=lambda x: max_length - len(x))
   word_list = [word.ljust(max_length, '_') for word in word_list]
   return word_list


print(equalize_lengths(['крот', 'белка', 'выхухоль']))
# ['выхухоль', 'белка___', 'крот____']

print(equalize_lengths(['a', 'aa', 'aaa', 'aaaa', 'aaaaa']))
# ['aaaaa', 'aaaa_', 'aaa__', 'aa___', 'a____']

print(equalize_lengths(['qweasdqweas', 'q', 'rteww', 'ewqqqqq']))

# ['qweasdqweas', 'ewqqqqq____', 'rteww______', 'q__________']
````


````python
def sum_min_numbers(a,b,c):
   num = [a,b,c]
   min_num = min(num)
   num.pop(num.index(min_num))
   min_num += min(num)
   return min_num

print(sum_min_numbers(1, 2, 3))
# 3
print(sum_min_numbers(1, 2, -10))
# -9
````

Примеры работы программы:

```
print(more_than_n([-1, 4, 4.2, 42.2, -3.4, -5.2], 3))
# [4, 4.2, 42.2, -3.4, -5.2]
print(more_than_n([-1, 4, 4.2, 42.2, -3.4, -5.2], 10))
# [42.2]
print(more_than_n([], 10))
# []
```

````python
def more_than_n(num,n):
   result = []
   for i in num:
      if abs(i) > n:
         result.append(i)
   return result


print(more_than_n([-1, 4, 4.2, 42.2, -3.4, -5.2], 3))
# [4, 4.2, 42.2, -3.4, -5.2]
print(more_than_n([-1, 4, 4.2, 42.2, -3.4, -5.2], 10))
# [42.2]
print(more_than_n([], 10))
# []
````

На вход функция принимает шестизначное число ticket_number и должна возвращать одно из булевых значений (True или False) в зависимости от того, является ли билет счастливым.

При решении постарайтесь не использовать встроенную функцию sum() — примените циклы.

**Примеры работы программы:**

```
print(lucky_ticket(111111))
# True
print(lucky_ticket(123456))
# False
```



````python
def lucky_ticket(n):
    n = str(n)

    first = n[0:3]
    second = n[3:6]

    sum_first = sum(map(int, first))
    sum_second = sum(map(int, second))

    return sum_first == sum_second



print(lucky_ticket(111111))
# True
print(lucky_ticket(123456))
# False
````

Медиана, как и среднее, является одной из важнейших характеристик для числовых данных в статистике.

Медиана — это число, которое находится в середине упорядоченной по возрастанию числовой последовательности.

Например, в последовательности чисел 1, 5, 2, 3, 6 медианой является число 3, так как если упорядочить числа по возрастанию (1, 2, 3, 5, 6), центральным элементом в последовательности будет именно 3.

Примечание. Если в выборке чётное число элементов, то за медиану принимается среднее между центральными элементами (после упорядочивания).

Например, в последовательности чисел 100, 5, 2, 4, 3, 6 медианой является число 4.5, так как, если упорядочить числа по возрастанию (2, 3, 4, 5, 6, 100), то центральным элементом в последовательности будет среднее арифметическое между 4 и 5 ( 4+52=4.5 ).

Пользователь с клавиатуры (через функцию input()) вводит через запятую и пробел неупорядоченную последовательность чисел.

Необходимо конвертировать ввод пользователя в список действительных чисел и вывести на экран медиану этой числовой последовательности.

Для поиска медианы в списке напишите функцию find_median(), которая будет возвращать одно число — медианное значение. Результат работы функции нужно вывести на экран.

Если переданная в функцию последовательность окажется пустой (пользователь ничего не ввёл) или пользователь ввёл числа некорректно (ввод нельзя преобразовать в числа), на экран должна быть выведена фраза Некорректный ввод.

**Пример №1:**

```
Ввод:
1, 5, 2, 3, 6

Вывод:
Median: 3.0
```

```
Ввод:
100, 5, 2, 4, 3, 6

Вывод:
Median: 4.5
```

```
Ввод:

Вывод:
Некорректный ввод
```

```
Ввод:
десять, 10, пять, 7, семь

Вывод:
Некорректный ввод
```

````python
numbers = input('Введите последовательность чисел: ')

def find_median(values):
    try:
        values = [int(x) for x in values.replace(',', '').split(' ')]
        values = sorted(values)
        n = len(values)
        mid = n // 2
        # нечётное количество
        if n % 2 == 1:
            return float(values[mid])

        # чётное количество
        return float((values[mid - 1] + values[mid]) / 2)
    except ValueError:
        return "Некоректный ввод"
find_median(numbers)
````

````python
input_string = input('Введите число словами: ')
number_word_dict = {
    "ты": 1000, "м": 1000000,
    "сто": 100, "двес": 200, "трис": 300, "четырес": 400, "пятьс": 500, "шестьс": 600, "семьс": 700, "восемьс": 800, "девятьс": 900,
    "одинн": 11, "двен": 12, "трин": 13, "четырн": 14, "пятн": 15, "шестн": 16, "семн": 17, "восемн": 18, "девятн": 19,
    "двад": 20, "трид": 30, "сор": 40, "пятьд": 50, "шестьд": 60, "семьд": 70, "восемьд": 80, "девяно": 90,
    "дес": 10, "н": 0, "о": 1, "дв": 2, "т": 3, "ч": 4, "п": 5, "ш": 6, "с": 7, "в": 8, "д": 9, }

def transform_string_to_integer(s,number_word_dict):
   # деление строки на слова для поиска по ключам
   words = s.split()

   total = 0 # глобальная сумма (ответ)
   current = 0 # локальня сумма (для вычесления)
   for word in words: 
      found_value = None #
      for key, value in number_word_dict.items(): # поиск значения по корню
         if word.startswith(key):
            found_value = value 
            break
      if found_value is None: continue # зашита от дурака
      if found_value < 1000: # для значений до 1000 (складываются в current)
         current += found_value
      elif found_value >= 1000: # если вдруг заданное число будет в тысячай стонях и тд, умнажается
         current = current * found_value
         total += current
         current = 0 # сброс current
   total += current
   return total

transform_string_to_integer(input_string, number_word_dict)
````

## Задание №5 (бонусная, не оценивается)
Разложить факториал на простые множители
Компания: «СБЕР»

Уровень: Сложный

Пользователь с клавиатуры (через функцию input()) вводит некоторое целое число n.

Напишите функцию decompose_factorial(), которая будет принимать на вход введённое пользователем число n, а затем вычислять факториал этого числа и выполнять разложение факториала на простые множители.

Напомним, что факториал числа  n  ( n! ) — это произведение всех натуральных чисел от  1  до  n  включительно.

Например, факториал числа  3  будет равен:  3!=1∗2∗3=6 .

Простое число — число, которое делится (нацело) только на себя и на 1. Например,  13  — простое число, так как оно делится только на себя (на  13 ) и на  1 , а  14  не является простым числом, так как оно делится на  7  и на  2 .

Подробнее о простых числах можно почитать здесь.

Простые множители — это простые числа (не считая 1), на которые число делится без остатка.

Например, для числа  10  простыми множителями являются числа  5  и  2 . Для числа  28  множителями являются числа  7  и  4 , но простыми множителями будут числа  7  и  2 , так как  4  — это  2∗2 , а число  28  представляется в виде  28=7∗22 .

В результате своей работы функция должна возвращать строку вида (приведён пример для n=12):

12! = 2^10 * 3^5 * 5^2 * 7 * 11

Здесь число слева от ^ — множитель, число справа от ^ — его степень. При этом если множитель встречается один раз, знак степени не выводится. Вывод множителей должен производиться в порядке их возрастания.

**Пример №1:**

```
Ввод:
5

Вывод:
5! = 2^3 * 3 * 5
Пример №2:

Ввод:
22

Вывод:
22! = 2^19 * 3^9 * 5^4 * 7^3 * 11^2 * 13 * 17 * 19
Пример №3:

Ввод:
25

Вывод:
25! = 2^22 * 3^10 * 5^6 * 7^3 * 11^2 * 13 * 17 * 19 * 23
```

**Подсказка №1:**

Функцию для вычисления факториала мы уже реализовывали, когда говорили о рекурсии в Python. Вы можете воспользоваться нашим вариантом функции

```
def factorial(n):
    if n == 1 or n == 0:
        return 1
    return n * factorial(n-1)
factorial(5)
или использовать реализацию из встроенной библиотеки math:

```

````python
input_num = input("Введите число: ")

def decompose_factorial(n):
   n = int(n)
   def find_factorial(n): # поиск факториала
      if n == 1 or n == 0:
         return 1
      return n * find_factorial(n-1)
   
   factorial = find_factorial(n) # факториал
   
   def factorize(num): # рвзлождение на простые множители
      d = 2 # начальный делитель
      factorz = {} # словарь с целыми делителми
      while num > 1: 
         while num % d == 0:
            factorz[d] = factorz.get(d,0) + 1 
            num //= d
         d += 1
      return factorz
   
   factorz = factorize(factorial)
   sorted_keys = sorted(factorz.keys()) # сортируем по ключам

   parts = [] # формируем словарь
   for key in sorted_keys:
      value = factorz[key]
      if key == 1:
         parts.append(str(key)) # что бы не выводить ^1
      else:
         parts.append(f'{key}^{value}') # добовляем значение степени
   result = f'{n}! = ' + '*'.join(parts) # формируем строку 
   return result

decompose_factorial(input_num)
````

## Задание №6 (бонусная, не оценивается)

#### Заменить дублирующиеся буквы

**Компания:** «СБЕР»

**Уровень:** Сложный

____

Пользователь с клавиатуры (через функцию `input()`) вводит строку, состоящую из символов латинского алфавита.

Необходимо последовательно заменять две одинаковые буквы во введённой строке на следующую по алфавиту букву (важно: заменяем по 2 одинаковых буквы за одну итерацию, не более). Операция замены должна повторяться до тех пор, пока в строке не останутся только уникальные символы.

**Несколько важных моментов:**

* буквы не обязаны быть смежными (смежные — стоящие рядом; в строке `'abc'` буквы a и b, а также b и c являются смежными, в то время как a и c — нет);
* zz заменяем на a;
* вводимая строка состоит из букв латинского алфавита в нижнем регистре.

Напишите функцию `replace_duplicates()`, которая принимает на вход произвольную строку `s`, значение которой вводится с клавиатуры. Функция должна последовательно производить описанную операцию и возвращать строку с оставшимися буквами в любом порядке.

**Пример:**
```
input:
zzzab

step 1: zaba
step 2: zbb
step 3: zc

output:
cz
```

```
input:
hhakafh

step 1: haakfi
step 2: hkfib

output:
hkfib
```

**Подсказка:**

#При решении задачи вам может понадобиться латинский алфавит. Вы можете получить его из встроенной в Python библиотеки string:

````python
import string
alphabet = string.ascii_lowercase

s = input('Введите строку: ')

def replace_duplicates(s,alphabet):
   counts = {}
   for ch in s: # подсчет частот
      counts[ch] = counts.get(ch, 0) + 1
   while True:
      changed = False
      for ch in alphabet:
         if counts.get(ch,0) >= 2:
            counts[ch] -= 2
            i = alphabet.index(ch)
            next_ch = alphabet[(i+1) % 26] 
            counts[next_ch] = counts.get(next_ch, 0)+1
            changed = True
            break
      if not changed:
         break
   result = ''
   for ch in alphabet:
      if counts.get(ch,0)>0:
         result += ch * counts[ch]
   return result

print(replace_duplicates(s,alphabet))
````
