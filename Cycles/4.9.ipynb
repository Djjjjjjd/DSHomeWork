Цикл **while** подразумевает немного другой подход к ограничению количества шагов, которые должны выполняться.

Цикл выполняется до тех пор, пока истинно его условие. Как только оно становится ложным, цикл прерывается.

Например, мы разрабатываем новую умную колонку «Алевтина». В ней заложен алгоритм воспроизведения любимой песни, пока пользователь не скажет одно из ключевых слов («Стоп», «Хватит», «Помолчи» и т. д.). Таким образом, здесь заложен цикл воспроизведения музыки, условие выхода из него — ключевое слово пользователя.

Для компьютера это выглядит как: «Делай, пока не наступит ...» .


Давайте рассмотрим структуру цикла while на примере.

Пусть мы хотим удалять элементы в списке lst=['a', 'b', 'c', 'd', 'e'], пока их не останется два.

Тогда синтаксис цикла для решения нашей задачи будет иметь следующий вид:


````python
lst = ['a','b','c','d','e']

while len(lst) > 2:
  lst.pop(0)
  print(lst)
  print('')
print(lst)
````

Числа Фибоначчи — это бесконечная последовательность чисел, которую можно получить рекурсивно. Каждое число из последовательности является суммой двух предыдущих.

Последовательность Фибоначчи активно применяется в комбинаторике и аналитике.

Напишите программу, которая выводит первые n чисел Фибоначчи, начиная с 1. Значение n может быть любым.


````python
n = 3
## fibonacci_list = [1, 1, 2]

n = 5
## fibonacci_list = [1, 1, 2, 3, 5]

n = 9
## fibonacci_list =[1, 1, 2, 3, 5, 8, 13, 21, 34]


# Инициализируем первые два числа Фибоначчи
a = 1
b = 1

# Создаем пустой список для хранения последовательности
fibonacci_list = []

# Если n равно 1, добавляем только первое число (1)
if n >= 1:
    fibonacci_list.append(a)
# Если n больше или равно 2, добавляем второе число (1)
if n >= 2:
    fibonacci_list.append(b)

# Начинаем с 2, потому что у нас уже есть 2 числа
i = 2

# Пока у нас меньше n чисел в списке
while i < n:
  # Вычисляем следующее число Фибоначчи (сумма двух предыдущих)
  next_fib = a + b
  # Добавляем новое число в список
  fibonacci_list.append(next_fib)
  # Обновляем значения a и b для следующей итерации
  # Старое значение b становится новым a
  a = b
  # Новое вычисленное число (next_fib) становится новым b
  b = next_fib
  # Увеличиваем счетчик чисел
  i += 1

print(fibonacci_list)
````

Такие структуры напоминают таблицу, и их ещё называют двумерными списками, а на математическом языке они называются двумерными матрицами.

Двумерная матрица представляет собой обычную таблицу. Если одномерная матрица — это список, то двумерная — это список списков.

> Для этого воспользуемся правилом «матрёшки»: если у нас есть вложенная структура, то для доступа к её элементам нужно раскрывать вложенность поэтапно. Открываем большую матрёшку, затем открываем матрешку поменьше и так далее, пока не дойдём до интересующего нас содержимого.

Давайте попробуем реализовать такой алгоритм для поиска минимумов.

Дана двумерная матрица 3x3 (список списков). Необходимо определить минимумы в каждой её строке.

random_matrix = [
[9, 2, 1],
[2, 5, 3],
[4, 8, 5]
]


````python
random_matrix = [
    [9, 2, 1],
    [2, 5, 3],
    [4, 8, 5]
]
# Задаём пустой список с минимальными значениями строк
min_value_rows = []
# Создаём цикл по строкам матрицы random_matrix
for row in random_matrix: # row — текущая строка таблицы
    # Задаём начальное значение кандидата на минимум
    min_value = row[0]
    # Создаём цикл по элементам списка row
    # elem — текущий элемент из списка row
    for elem in row:
        # Проверяем условие, что текущий элемент меньше кандидата на минимум.
        if elem < min_value:
            # Если условие выполняется, заменяем кандидата на минимум.
            min_value = elem
    # Добавляем полученный минимум строки в список
    min_value_rows.append(min_value)
# Выводим минимальные элементы
print("Minimal elements:", min_value_rows)
````

Что такое абсолютное значение?

Абсолютное значение числа — это его расстояние от нуля на числовой прямой, независимо от его знака.

Для положительных чисел абсолютное значение равно самому числу. Например, abs(5) равно 5.
Для отрицательных чисел абсолютное значение равно тому же числу, но без знака минус. Например, abs(-5) равно 5.
Для нуля абсолютное значение равно нулю. Например, abs(0) равно 0.

Необходимо написать программу, которая проверяет, что число n является степенью числа 3.


````python
n = 27
count = 0
while True:
  if n % 3 == 0:
    n = n // 3
    count += 1
    if n == 1:
      print(f'Число является степенью 3x, число итераций {count}')
      count = 0
      break
  else:
      print(f'Число не является степенью 3x')
      break
````

Допишите программу, которая проверяет гипотезу Сиракуз.

Гипотеза Сиракуз заключается в том, что любое натуральное число n можно свести к 1, если повторять над ним следующие действия:

если число чётное, разделить его нацело пополам, т. е. n = n // 2;
если нечётное — умножить на 3, прибавить 1 и результат нацело разделить на 2, т. е. n = (n \* 3 + 1) // 2.

**Мы тестируем только целые числа!**

Пример работы:

```
n = 19
## Syracuse hypothesis holds for number 19

n = 25921
## Syracuse hypothesis holds for number 25921

```


````python
n = 19
## Syracuse hypothesis holds for number 19

# n = 25921
## Syracuse hypothesis holds for number 25921

# Мы тестируем только целые числа!

while True:
  if n == 1:
    break
  if n % 2 == 0:
    n = n // 2
  else:
    n = (n * 3 + 1) // 2

# Проверяем, что результат равен 1
print(f'Syracuse hypothesis holds for number: {n}')
````
