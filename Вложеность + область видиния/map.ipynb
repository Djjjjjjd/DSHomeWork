# MAP

map() — это функция в Python, которая применяет другую функцию ко всем элементам списка (или любого набора данных) и возвращает новый набор результатов.

Это как мини-конвейер:
берёт элементы → пропускает через функцию → отдаёт обработанные элементы.

```
map(функция, коллекция)

```

функция — что делать с каждым элементом

коллекция — список, кортеж и т.д.

Обычно после map() сразу делают list(), чтобы увидеть результат:

```
list(map(функция, коллекция))

```

Как это работает (простыми словами)

Ты создаёшь функцию, которая умеет обрабатывать один элемент.

map() прогоняет каждый элемент списка через эту функцию.

Получаешь новый список результатов.



Простой пример

Допустим, нужно умножить каждый элемент списка на 2.

Без map():
```
nums = [1, 2, 3, 4]
result = []

for n in nums:
    result.append(n * 2)

print(result)  # [2, 4, 6, 8]

```

C map():
```
nums = [1, 2, 3, 4]

def multiply_by_two(x):
    return x * 2

result = list(map(multiply_by_two, nums))

print(result)  # [2, 4, 6, 8]
```

Очень короткий вариант (lambda):
```
nums = [1, 2, 3, 4]
result = list(map(lambda x: x * 2, nums))
print(result)  # [2, 4, 6, 8]
```

````python
docs = [  
'//doc/5041434?query=data%20science',  
'//doc/5041567?query=data%20science',  
'//doc/4283670?query=data%20science',  
'//doc/3712659?query=data%20science',  
'//doc/4997267?query=data%20science',  
'//doc/4372673?query=data%20science',  
'//doc/3779060?query=data%20science',  
'//doc/3495410?query=data%20science',  
'//doc/4308832?query=data%20science',  
'//doc/4079881?query=data%20science'  
]

def addLink(link):
    def adlink(link):
        return 'https://www.kommersant.ru' + link
    links = list(map(adlink, link))
    return links
links = addLink(docs)

````
